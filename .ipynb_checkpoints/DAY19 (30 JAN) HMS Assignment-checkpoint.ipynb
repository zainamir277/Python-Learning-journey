# HMS

```python
import sqlite3
```

```python
def connect():
    return sqlite3.connect("assignment 30Jan HMS")
```

```python
table_query="""
create table patients(
id integer primary key AUTOINCREMENT,
name text,
address text,
phone text,
doctor text,
visit_date text,
remarks text
)
"""
con=connect()
con.execute(table_query)
```

# Add Patient:

```python
import datetime
```

```python
def addpatient():
    nm=input("Enter Patient's Name : ")
    ph=input("Enter Patient's Phone Number : ")
    ad=input("Enter Patient's Address : ")
    dr=input("Enter Doctor Name : ")
    vd=str(datetime.date.today())
    rmk=input("Enter Doctor's Remarks : ")
    con=connect()
    cursor=con.cursor()
    cursor.execute("""
    insert into patients(name, address, phone, doctor, visit_date, remarks) Values (?,?,?,?,?,?)""",(nm, ad, ph, dr, vd, rmk))
    con.commit()
    con.close()
```

```python
addpatient()
```

# Remove Patient :

```python
def removepatient():
    ph=input("Enter Patient's phone number :")
    con=connect()
    cursor=con.cursor()
    cursor.execute("delete from patients where phone=?",(ph,))
    con.commit()
    if cursor.rowcount > 0:
        print("Patient's data has been sucessfully removed")
    else:
        print("No patient found")
    con.close()
```

```python
removepatient()
```

## Update Patient Data:

```python
def updatepatientrecord():
    phone=input("Enter Patient's previous phone number : ")
    nm=input("Enter Patient's Name : ")
    ph=input("Enter Patient's Phone Number : ")
    ad=input("Enter Patient's Address : ")
    dr=input("Enter Doctor Name : ")
    vd=str(datetime.date.today())
    rmk=input("Enter Doctor's Remarks : ")
    con=connect()
    cursor=con.cursor()
    cursor.execute("Update patients set name=?, address=?, phone=?, doctor=?, visit_date=?, remarks=? where phone=?",(nm, ad, ph, dr, vd, rmk, phone))
    con.commit()
    con.close()
    print("Patient's record has been sucessfully updated")
```

```python
updatepatientrecord()
```
