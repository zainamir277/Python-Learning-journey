# A Python program: DataBase of hospital management system

### Database of hospital management system on sqlite3 containing Patient name, address, phone, doctor name, visit date, and remarks. ADD/Remove/Update Patient details

```python
# import sqlite3 to make database
import sqlite3
```

```python
# Make a connect function to connect sqlite3 again and again.
def connect():
    return sqlite3.connect("hmsgh.db")
```

```python
# Make a table that contain Patient name, address, Contact, Doctor name, visit date, remarks 

table_query="""
create table patients(
id integer primary key AUTOINCREMENT,
patient text,
address text,
phone text,
doctor text,
visit_date text,
remarks text
)
"""
con=connect()
con.execute(table_query)
```

```python
# import datetime to take current date
import datetime
```

```python
# Make a function to add patient
def addpatient():
    nm=input("Enter Patient Name : ")
    ad=input("Enter Patient Address : ")
    ph=input("Enter Patient Phone Number : ")
    dr=input("Enter Doctor Name : ")
    vd=str(datetime.date.today())
    rmk=input("Enter Doctor Remarks : ")
    con=connect()
    cursor=con.cursor()
    cursor.execute("insert into patients(patient, address, phone, doctor, visit_date, remarks) Values(?,?,?,?,?,?)",(nm, ad, ph ,dr, vd, rmk))
    con.commit()
    con.close()
```

```python
# Make function to remove Patient
def removepatient():
    ph=input("Enter Patient's phone number :")
    con=connect()
    cursor=con.cursor()
    cursor.execute("delete from patients where phone=?",(ph,))
    con.commit()
    if cursor.rowcount > 0:
        print("Patient's data has been sucessfully removed")
    else:
        print("No patient found")
    con.close()
```

```python
# Make function to update patient details:
def updatepatientdata():
    oldphone=input("Enter Patient Prevous Numeber : ")
    nm=input("Enter Patient Name : ")
    ad=input("Enter Patient Address : ")
    ph=input("Enter Patient Phone Number : ")
    dr=input("Enter Doctor Name : ")
    vd=str(datetime.date.today())
    rmk=input("Enter Doctor rmks")
    con=connect()
    cursor=con.cursor()
    cursor.execute("Update patients set patient=?, address=?, phone=?, doctor=?, visit_date=?, remarks=? where phone=?",(nm, ad, ph, dr, vd, rmk, oldphone))
    con.commit()
    con.close()
```
